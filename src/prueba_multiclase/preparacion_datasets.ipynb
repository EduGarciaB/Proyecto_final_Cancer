```python
df3 = pd.read_csv ('/Users/luiseduardogarciablanco/Desktop/nueva data cancer/2020/ISIC_2020_Training_GroundTruth (1).csv', index_col=0)
df3
```

```python
#para preparar este dataset para igualarlo y sumarlo a los anteriores
df3 = df3[['diagnosis']]
df3
```

```python
#revisamos cuentos valores unicos tiene la columna diagnosis
# Contar el número de valores únicos
num_valores_unicos = df3['diagnosis'].nunique()

# Obtener los valores únicos
valores_unicos = df3['diagnosis'].unique()

# Mostrar los resultados
print(f'Número de valores únicos en la columna "diagnosis": {num_valores_unicos}')
print('Valores únicos en la columna "diagnosis":')
print(valores_unicos)
```

```python
# Realizar el one-hot encoding
df3_encoded = pd.get_dummies(df3['diagnosis'], prefix='diagnosis')

# Concatenar las columnas codificadas al DataFrame original
df3 = pd.concat([df3, df3_encoded], axis=1)

# Opcional: eliminar la columna original si ya no la necesitas
df3.drop('diagnosis', axis=1, inplace=True)
df3
```

```python
# Eliminar el prefijo "diagnosis_" de las columnas
df3.columns = df3.columns.str.replace('diagnosis_', '', regex=False)
df3
```

```python
# Renombrar las columnas específicas
df3.rename(columns={'melanoma': 'MEL', 'nevus': 'NV'}, inplace=True)
df3
```

```python
# Convertir valores booleanos a enteros (1 y 0)
df3 = df3.astype(int)
df3
```

```python
#concatenamos los dataframe en uno nuevo
df_unificado = pd.concat([ df1, df2, df3])
df_unificado
```

```python
#cambiamos valores Nan por 0.0 donde unico los hay que es en la columna SCC ya que no corresponde a ninguna variable
df_unificado.fillna(0.0, inplace=True)
df_unificado
```

```python
#binarizamos las columnas para que sea mas facil el aprendizaje de los modelos
df_unificado = df_unificado.astype(int)
df_unificado
```
