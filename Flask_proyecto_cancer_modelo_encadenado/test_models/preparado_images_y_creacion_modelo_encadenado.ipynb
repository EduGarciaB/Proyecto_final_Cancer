```python
import os
import pandas as pd

# Cargar el dataset (asegúrate de cargarlo correctamente)
df = pd.read_csv('/Users/luiseduardogarciablanco/Desktop/bootcamp/Flask_proyecto_cancer_jpg_data/static/data/dataset_flask.csv')  # Reemplaza con la ruta a tu dataset
image_folder = '/Users/luiseduardogarciablanco/Desktop/nueva data cancer/test web/jpg_ensambled_model'  # Ruta a la carpeta de imágenes

# Asegúrate de que la columna 'isic_id' tenga la extensión .jpg
df['isic_id'] = df['isic_id'].apply(lambda x: f"{x}.jpg")

# Obtener una lista de los nombres de los archivos en la carpeta de imágenes
image_files = set(os.listdir(image_folder))

# 1. Eliminar imágenes que no estén en el dataset
for image_file in image_files:
    # Si el archivo no está en el dataset, se elimina
    if image_file not in df['isic_id'].values:
        file_path = os.path.join(image_folder, image_file)
        print(f"Borrando: {file_path}")
        os.remove(file_path)

# 2. Comprobar si hay registros en el dataset que no tengan una imagen correspondiente en la carpeta
missing_images = []
for image_id in df['isic_id']:
    if image_id not in image_files:
        missing_images.append(image_id)

# Mostrar los registros que no tienen una imagen correspondiente
if missing_images:
    print("Registros en el dataset sin imagen correspondiente:")
    print(missing_images)
else:
    print("Todas las imágenes están presentes para los registros del dataset.")
```
